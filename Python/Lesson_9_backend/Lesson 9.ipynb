# Backend-разработка

## Что это такое
Под backend-разработкой обычно подразумевают разработку серверной части в приложениях, состоящих из интерфейса для пользователя и серверной логики.

Примеры:
1. При разработке сайтов в качестве backend-сервера выступает компьютер, который обрабатывает запросы пользователя и хранит информацию о пользователях. Backend-разработка в таком случае - это разработка софта для такого сервера.
2. При разработке мобильного приложения часто делается сервер, хранящий данные пользователя и раздающих их в том или ином виде. Backend-разработка здесь включает в себя то же, что и в случае сайта.

## Зачем это нам
Как только ML-модель готова, возникает задача донести ее до продакшена и научиться получать ее предсказания в любой момент. Для этих целей хорошо подходит веб-сервис: модель сохраняется на сервер и там же запускается веб-сервис.

Пользователь отправляет запрос на веб-сервис, сервис запускает предсказания модели и отдает обратно пользователю результат.

# Основы веба: запрос
Большая часть того, что вы видите в браузере, работает на механизме _веб-запросов_ (коротко _requests_).

<div>
<img src="https://miro.medium.com/max/1066/1*dD0vX4r4QreNOhG3gqDjmg.png" width="900"/>
</div>

**Запрос** - это обращение к серверу с просьбой вернуть некие данные. Обращение может включать в себя данные, необходимые для его обработки, а может не включать.

Ожидается, что сервер ответит на каждый запрос (даже если никакой значимой информации передавать не надо) - хотя бы чтобы дать знать, что все ок и запрос обработан.

_Веб-запрос_ включает включает в себя несколько параметров:
- путь
- метод запроса
- query params
- заголовки запроса
- тело запроса (не везде)

Ответ от сервера включает те же параметры, но добавляет еще свои:
- status code (все ли прошло хорошо или есть проблемы)
- информацию о сервере
- загловки по безопасности

Мы разберем все эти параметры подробнее.

_Запрос выглядит как строка, которая формируется специальным образом и потом отправляется на сервер._

</div>

## Путь
Путь - это все, что идет в адресной строке после отсечения протокола и адреса сайта. Например, для `https://lab.karpov.courses/learning/48/module/618/lesson/5126/14946/59186/` путем будет `/learning/48/module/618/lesson/5126/14946/59186/`.

Путь нужен серверу для того, чтобы ориентироваться среди доступных страниц/ресурсов. Это похоже на навигацию по папкам: путь на сайте в этом случае будет играть схожую роль, что и путь к папке.

## Методы: GET, POST, PUT
Есть в каждом запросе. Говорят серверу основную цель запроса:
1. GET - это получить некоторые данные. Например, при загрузке страницы пользователя в соц.сети будет отправлен GET запрос на его основные данные.
2. PUT - загрузить новые данные на сервер. Например, при регистрации пользователя может отправиться PUT запрос с регистрационными данными.
3. POST - отредактировать информацию, либо передать некоторую сущность на сервер. Допустимы побочные эффекты (например, изменение состояния сервера, обновление данных в БД).
На деле методов больше, но GET, PUT, POST используются чаще всего.

[Подробнее про методы](https://developer.mozilla.org/ru/docs/Web/HTTP/Methods)

## Query params
Передаются в адресной строке. По сути - способ передать данные в форме "ключ=значение". Синтаксис - вопросительный знак, потом "ключ=значение":
```
/user?id=5&limit=10
/?q=hello
```

## Заголовки запроса
Содержат служебную информацию: какой формат данных ожидается, какой браузер клиента (нестрого говоря), какой язык у клиента и т.п.

По-английски называются _requests headers_. Часто на жаргоне зовутся "хидерами".

### Cookie
Кстати, в заголовках передаются cookie. Cookie - это небольшая порция данных, который сервер может сохранить у клиента и получать обратно при каждом запросе.
Cookie хранятся в формате `{ключ: значение}`, как словари в Python.

Куки используются для авторизации (устаревший способ, но еще весьма популярный), идентификации одного и того же браузера (для рекламы это актуально).

Cookie имеют срок жизни, после которых они считаются невалидными. Кроме того, клиент может эти куки удалить в любой момент (в браузере это кнопка "Почистить куки").

Пример headers и cookie:

```
GET /sample_page.html HTTP/2.0
Host: www.example.org
Cookie: yummy_cookie=choco; tasty_cookie=strawberry
```

## Тело запроса
После указания метода и хидеров можно перейти к главной части - к _телу запроса_. В теле запроса передается любая информация в любом виде.

Мы будем работать со случаями, когда в тело будет уходить JSON. JSON - это формат, схожий со словарем Python.

Пример JSON:
```json
{"menu": {
  "id": "file",
  "value": "File",
  "popup": {
    "menuitem": [
      {"value": "New", "onclick": "CreateNewDoc()"},
      {"value": "Open", "onclick": "OpenDoc()"},
      {"value": "Close", "onclick": "CloseDoc()"}
    ]
  }
}}
```

Преимущество такого формата - его легко превратить в единую строку и положить в тело запроса.

Сервер сам решает, как поступать с этой информацией. Но есть правило: **в GET запросе тело обычно пусто**. Передавать ее можно, но это будет плохой практикой.

Соответственно, мы будем передавать _тело запроса_ только в POST и PUT запросах.

# Основы веба: ответ
После получения запроса сервер обрабатывает его каким-то образом, затем обязательно возвращает _ответ_ (_server response_).

В ответе идут те вещи, что в запросе (кроме параметров) и некоторые дополнительные. О них и поговорим дальше.

## Status code
Число, несущее информацию о результатах обработки. Расшифровка кодов [тут](https://developer.mozilla.org/ru/docs/Web/HTTP/Status).

Самые частые:
- 200 - все ок
- 30* - будет перенаправление
- 400 - запрос плохой и не будет обработан. 
- 401 - неавторизован. Скорее всего, нужно логиниться.
- 403 - запрещено. Вы залогинены, но у вас нет прав на это.
- 404 - не найдено. Вы идете по пути, которого нет у сервера. Либо запрашиваете объект, который не найден на сервере. Иногда объект может быть, но прав нет - и сервер решит скрыть его наличие.
- 500 - сервер упал. Плохой знак для разработчика - у вас есть необработанная ошибка.
- 502, 504 - проблемы шлюза. Обычно возникают, когда на сервере внутри проблемы с внутренней коммуникацией компонент.

Запомнить их легко по мнемонике:
- 20* - все хорошо.
- 30* - технические коды по процессу соединения с сервером.
- 40* - "клиент неправ" (ошибка на стороне клиента)
- 50* - "сам дурак" (ошибка на стороне сервера)

Нам понадобятся эти знания, когда будем писать свои запросы.

## Информация о сервере
Любая инфа, которую сервер захочет сообщить:
- версия ПО
- версия ОС
- время на сервере
- и т.п.

## Заголовки по безопасности
Есть интересные трюки по вытягиванию информации с одного сайта на другом сайте.

Один из них - CSRF-атака. Не будем вдаваться в детали, но скажем, что для защиты сервер передает т.н. _CSRF-токен_. Вам необходимо будет уважать это решение и аккуратно его обходить. Современные веб-фреймворки дают это сделать.

[Описание CSRF из вики с примером](https://ru.wikipedia.org/wiki/Межсайтовая_подделка_запроса).

# Практика
## FastAPI
Веб-фреймворк для быстрого создания производительных веб-приложений. Прост в освоении (по сравнению с Django) и хорошо подходит для написания REST API.

## REST API
API расшифровывается как Application Programming Interface. API "говорит" о том, какие рычаги для взаимодействия открывает наше приложение сторонним разработчикам.

К примеру, мы можем разрабатывать программу, храняющую исторую платежей пользователя. Этой программе можно добавить возможность отдавать историю конкретного пользователя: например, через веб-запрос. В таком случае говорят, что есть API для получения истории конкретного пользователя.

REST - это аббревиатура для REpresentational State Transfer. Звучит сложно, но на практике это означает следующее:
- Все взаимодействие программ строится по модели "сервер-клиент". Т.е. кто-то выступает в роли сервера, обрабатывая запросы, и второе лицо выступает в роли клиента, отправляя запросы.
- Отсутствие состояния. Между двумя запросами сервер не хранит состояния пользователя. Все необходимые для обработки запроса данные пользователь должен передавать сам.
- Сервер выставляет наружу понятные пути, объединяя их логически в блоки, и активно использует _методы запроса_. К примеру, для операций со списком пользователей будут следующие URL:
  - GET /user/\<id\> (выгрузить)
  - GET /user/me (выгрузить себя)
  - POST /user/me (обновить себя)
  - POST /user/ (добавлять произвольного)
